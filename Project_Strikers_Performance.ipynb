```python
from scipy.stats import stats
from scipy.stats import f_oneway
spain_var = df.query("Nationality == 'Spain'")['Consistency']
france_var =  df.query("Nationality == 'France'")['Consistency']
germany_var =  df.query("Nationality == 'Germany'")['Consistency']
brazil_var =  df.query("Nationality == 'Brazil'")['Consistency']
england_var =  df.query("Nationality == 'England'")['Consistency']
Test_stat, p_value = stats.f_oneway(spain_var, france_var, germany_var, brazil_var, england_var)
print("P-Value:",p_value)
if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
from scipy.stats import levene
stats, p_value = levene(spain_var, france_var, germany_var, brazil_var, england_var)
print("P value:", p_value)
```

```python
df['Total contribution score'] = df['Goals Scored']+ df['Assists'] + df['Shots on Target']+ df['Dribbling Success']+df['Aerial Duels Won']+ df['Defensive Contribution']+ df['Big Game Performance']+ df['Consistency']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['encoded_feature'] = encoder.fit_transform(df['Footedness'])
df['encoded_feature1'] = encoder.fit_transform(df['Marital Status'])
df.head()
```

```python
dummies = pd.get_dummies(df[['Nationality']])
df=pd.concat([df, dummies],axis = 1)
df.head()
```

```python
x=df.drop(['Striker_ID',],axis =1)
```

```python
x = pd.get_dummies(x, drop_first=True)
```

from sklearn.cluste import KMeans

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init = 'k-means++')
    kmeans.fit(x)
    wcss_score = kmeans.inertia_
    wcss.append(wcss_score)
wcss
```

```python
plt.figure(figsize = (6,3))
plt.plot(range(1,11),wcss,marker = 'o')
plt.title('elbow methods')
plt.xlabel('no of clusters- k')
plt.ylabel('wcss')
plt.show()
```

```python
final_km = KMeans(n_clusters = 2)
final_km.fit(x)
labels = final_km.labels_
labels
```

```python
df['Clusters']=labels
df.head()
```

```python
round(df.groupby('Clusters')[['Total contribution score']].mean(),2)
```

```python
mapping = {0:'Regular Striker',1:'Best Striker'}
df['Strikers types'] = df['Clusters'].map(mapping)
df.head()
```

```python
x=df.drop(['Striker_ID'],axis =1)
x
```

```python
from sklearn.preprocessing import StandardScaler
y = df['Strikers types']
x = pd.get_dummies(x, drop_first=True)
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_reg_train, x_reg_test,y_reg_train,y_reg_test = train_test_split(scaled,y,test_size=0.2,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
model=LogisticRegression()
model.fit(x_reg_train, y_reg_train)
y_pred = model.predict(x_reg_test)
accuracy = accuracy_score(y_reg_test, y_pred)
print("A:",accuracy)
```

```python
conf_matrix = confusion_matrix(y_reg_test, y_pred)
plt.figure(figsize = (6,3))
sns.heatmap(conf_matrix,annot = True, fmt='d', cmap = "Blues")
plt.title('CM')
plt.xlabel('Pred')
plt.ylabel('Actual')
plt.show()
```
